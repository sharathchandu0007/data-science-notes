# Stack Plotting

* Datasets stacks one on the other

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Dataset

```python
import pandas as pd
import numpy as np

x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

y = np.vstack([y1,y2,y3])
y
```

### Basic Stack Plotting 

```python
fig1, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.stackplot(x, y1, y2, y3,)
ax1.legend(loc=2)
ax2.stackplot(x, y)
ax2.legend(loc=2)
plt.show()
```

### baseline

* 'zero': Constant zero baseline, i.e. a simple stacked plot.
* 'sym': Symmetric around zero and is sometimes called ‘ThemeRiver’.
* 'wiggle': Minimizes the sum of the squared slopes.
* 'weighted_wiggle': Does the same but weights to account for size of each layer. It is also called ‘Streamgraph’-layout. More details can be found at http://leebyron.com/streamgraph/.

```python
plt.stackplot(x,y,baseline='zero');
```

```python
plt.stackplot(x,y,baseline='sym');
```

```python
plt.stackplot(x,y,baseline='wiggle');
```

```python
plt.stackplot(x,y,baseline='weighted_wiggle');
```

### labels

* Takes Length N sequence of strings
* Labels to assign to each data series

```python
labels = ["Fibonacci ", "Evens", "Odds"]
plt.stackplot(x,y,labels=labels,baseline='zero')
plt.legend(loc=2)
plt.show();
```

### colors

* Colors the stacked areas
* Takes a list or tuple of colors


```python
plt.stackplot(x,y,labels=labels,baseline='zero',colors=['green','white','orange'])
plt.legend(loc=2)
plt.show();
```

## Plot on DataFrame
* Basic Plotting 
 *  x, y, Takes label or position value
 *  These are optional parameters

```python
import pandas as pd

data = pd.DataFrame({'x':x, 'y1':y1 ,'y2':y2, 'y3':y3})
data
```

```python
data.plot.area();
```

```python
data.plot.area(x = 'x', y = ['y1','y2', 'y3']);
```

* Along with x and y plot also can take keyword arguments to pass on to pandas.DataFrame.plot()

```python
data.plot.area(x = 'x', y = ['y1','y2', 'y3'],stacked = False);
```
