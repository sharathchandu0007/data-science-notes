## Basic Plotting

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'X': np.arange(10)})
df['X^2'] = df.X**2
df
```

* Plotting Data in DataFrame df

```python
plt.plot(df['X'],df['X^2'])
```

* ploting using dataframe plot

```python
df.plot()
```

## Legends, Title, Labels, Xticks,Yticks,Color,Linewidth,Colormap, Grid, Kind

```python
title  :
xlabel :
ylabel :
legend :
xticks :
yticks :
colormap:
grid :
kind :
```

### To set title ,labels,legend,ticks

```python
plt.plot(df['X'],df['X^2'])
plt.title('X vs X^2 Plot')
plt.xlabel('X')
plt.ylabel('X^2')
plt.legend()
plt.xticks([2,4,6,8])
plt.yticks([4,16,36,64])
plt.show()
```

```python
df.plot(title='X vs X^2 Plot',legend=True,xticks=[2,4,6,8],yticks=[4,16,36,64],x='X',y='X^2')
```

* Plotting data when dataframe having more than 2 columns

```python
df['X^3'] = df.X**3
df
```

```python
df.plot()
```

```python
plt.plot(df['X'],df['X'])
plt.plot(df['X'],df['X^2'])
plt.plot(df['X'],df['X^3'])

plt.legend(loc='center left', bbox_to_anchor=(1,0.2))
plt.xlabel('X')
plt.ylabel('Result')
```

* Color and Linewidth 

```python
df.plot(title='Plottting Powers of X', color = ['b','g','c'],linewidth= 5)
```

PS: List of linewidth will not work

```python
plt.plot(df['X'],df['X'],c='b',label= 'X line',linewidth=5)
plt.plot(df['X'],df['X^2'],c='g',label= 'X^2 line',linewidth=4)
plt.plot(df['X'],df['X^3'],c='c',label= 'X^3 line', linewidth=3)

plt.title('Plottting Powers of X')
plt.legend()
```

* Colormap

```python
df.plot(title='Plottting Powers of X', colormap='winter')
```

```python
plt.plot(df['X'],df['X'],label= 'X line',linewidth=5)
plt.plot(df['X'],df['X^2'],label= 'X^2 line',linewidth=4)
plt.plot(df['X'],df['X^3'],label= 'X^3 line', linewidth=3)

plt.title('Plottting Powers of X')
plt.set_cmap('winter')
plt.show()
```

colarmap values: <a href ='https://matplotlib.org/examples/color/colormaps_reference.html' > </a>

* Grid

```python
plt.plot(df['X'],df['X'])
plt.plot(df['X'],df['X^2'])
plt.plot(df['X'],df['X^3'])

plt.title('Plottting Powers of X')
plt.xlabel('X')
plt.ylabel('Rsult')
plt.legend()
plt.grid(True,color='m')
plt.show()
```

```python
df.plot(title='Plottting Powers of X', colormap='winter',grid=True)
```
