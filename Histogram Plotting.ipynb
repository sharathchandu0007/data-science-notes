### bins

* Can configure the number of bins
* Takes integer or sequence or auto
* If an integer is given, bins + 1 bin edges are calculated and returned, consistent with numpy.histogram().
* If bins is a sequence, gives bin edges, including left edge of first bin and right edge of last bin. In this case, bins is returned unmodified <br/>
  All but the last (righthand-most) bin is half-open. In other words, if bins is:
  [1, 2, 3, 4]<br/>
  then the first bin is [1, 2) (including 1, but excluding 2) and the second [2, 3). The last bin, however, is [3, 4], which  includes 4.<br/>
  Unequally spaced bins are supported if bins is a sequence.
* If Numpy 1.11 is installed, may also be 'auto'.

* Default is taken from the rcParam hist.bins.


```python
plt.hist([data['a'],data['b'],data['c']], bins=5);
```

```python
plt.hist([data['a'],data['b'],data['c']], bins=25);
```

### range

* Can configure lower and upper range of the bins
* Takes tuple or None
* Points beyond the range are considered as outliers and ignores them
* If not provided, range is (x.min(), x.max())
* Range has no effect if bins is a sequence
* If bins is a sequence or range is specified, autoscaling is based on the specified bin range instead of the range of x.
* By default the value sets to None

```python
plt.hist([data['a'],data['b'],data['c']],bins=4,range=(-2, 2));
```

```python
plt.hist([data['a'],data['b'],data['c']],bins=[1,2,3,4],range=(-2, 2));
```

### cumulative

* If True, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values.
* The last bin gives the total number of datapoints. 
* If normed or density is also True then the histogram is normalized such that the last bin equals 1.
* If cumulative evaluates to less than 0 (e.g., -1), the direction of accumulation is reversed. 
* In this case, if normed and/or density is also True, then the histogram is normalized such that the first bin equals 1.
* By default the value sets to False


```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)),cumulative=True);
```

### bottom

* Takes location of the bottom baseline of each bin
* Takes array_like, scalar, or None value
* If a scalar, the base line for each bin is shifted by the same amount
* If an array, each bin is shifted independently and the length of bottom must match the number of bins
* If None, defaults to 0
* By default the value sets to None

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)),cumulative= True, bottom=0.2);
```

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True, weights=np.random.randint(1,100,size=(data.shape[0],3)),
         cumulative= True, bottom=[0.0,0.0,0.2,0.22,0.0,0.0,0.0,0.2,0.0,0.25]);
```

### histtype

* Can specify the type of histogram to draw
* Takes 'bar' | 'barstacked' | 'step' | 'stepfilled' value
* If bar, bars are arranged side by side
* If barstacked, multiple data are stacked on top of each other
* If step, generates a lineplot that is by default unfilled.
* If stepfilled, generates a lineplot that is by default filled
* By default the value sets to 'bar'

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True, weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,
         histtype='step');
```

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True, weights=np.random.randint(1,100,size=(data.shape[0],3)),
         bottom=0.0,histtype='bar');
```

### align

* Can controls how the histogram is plotted.
* Takes 'left' | 'mid' | 'right' value
* If left,  bars are centered on the left bin edges
* If right, bars are centered on the right bin edges
* If mid, bars are centered between the bin edges
* By default the value sets to mid

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)),bottom=0.0, histtype='bar', align='left');
```

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,histtype='bar', align='right');
```

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,histtype='bar', align='mid');
```

### rwidth

* Can specify the  relative width of the bars as a fraction of the bin width
* Takes Scalar or None vlaue
* If None, automatically compute the width.
* If histtype sets to step, or stepfilled , ignores the rwidth specified
* By default the vlaue sets to None

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True, 
         weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,histtype='bar', align='mid', rwidth= 1.0);
```

### log

* If True, the histogram axis will be set to a log scale. 
* If log is True and x is a 1D array, empty bins will be filtered out and only the non-empty (n, bins, patches) will be returned.
* By default the value sets to False

```python
n, bins,patchs = plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
                          weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,histtype='bar', align='mid',
                          rwidth=1.0,log=True);
```

```python
print('n:\n', n)
print('bins:\n', bins)
print('patchs\n', patchs)
```

### color

* Provides color to datasets(one color for dataset)
*  Takes color or array_like of colors or None value
* By default the vlue sets to None i.e standard line color sequence


```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,histtype='bar', align='mid', rwidth=1.0,log=True,
         color=['g','y','r']);
```

### label

* Can provide label information to plot
* Takes String, or Sequence of strings to match multiple datasets
* If label is not None, legend can  consumes the label values
* By default the value sets to None 

```python
plt.hist([data['a'],data['b'],data['c']],range=(x.min(),x.max()), density=True,
         weights=np.random.randint(1,100,size=(data.shape[0],3)), bottom=0.0,histtype='bar', align='mid', rwidth=1,
         color=['g','y','r'],label=['a','b','c']);
plt.legend()
plt.show()
```

### stacked

* If True, multiple data are stacked on top of each other
* If False multiple data are arranged side by side if histtype is bar or on top of each other if histtype is step
