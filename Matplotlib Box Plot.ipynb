# Box- whisker Plot

* Make a box and whisker plot for each column of x or each vector in sequence x., wiht a line at the median
* It is way of representing statistical data on a plot in which a rectangle is drawn to represent the second(Q2) and third(Q3) quartiles, usually with a vertical line inside to indicate the median value. The lower(Q1) and uppe(Q4)r quartiles are shown as horizontal lines either side of the rectangle.
* The whiskers(lower and Upper vertical lines) from the box to show the range of the data
* Flier(outlier) points are those are on above or below the whiskers.

### Dataset

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('../Advertising.csv',index_col= 0)
```

```python
data.head()
```

### Basic Plotting 

```python
plt.boxplot(data['TV']);
```

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']]);
```

### notch

* The notches represent the confidence interval (CI) around the median 
* CI ,In statistics, a confidence interval is a type of interval estimate that is computed from the observed data. 
  The confidence level is the frequency of possible confidence intervals that contain the true value of their 
  corresponding parameter.
* If True, will produce a notched box plot. Otherwise, a rectangular boxplot is produced
* Takes boolean value(True/False)
* By default the value sets to False

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']], notch=True);
```

### sym

* Provides symbol for outlier points
* By default the symbol is set to 'b+'(shape o)
* Takes String value
* To not to show any symbol pass empty string to sym

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']], sym='');
```

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']], sym='x');
```

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None);
```

### vert

* Can specify the orientation of graph
* Takes bool value(True/False)
* If True, makes the boxes verical. If else, Horizontal
* By default the value sets to True

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=False);
```

### whis

* Determines the reach of the whiskers to the beyond the first and third quartiles
* IQR or Box --> Q3-Q1, <br/> 
  Upper Whisker or Upperr line --> Q3 + whis* IQR
  Lower Whisker or lower line -->  Q1 - Whis* IQR
* Beyond the Whisker/line, data is considered outlier and are plotted as individual points(by default)
* Takes float or 'range' value
* If range , whisker covers all the data points in Q1 and Q4
* By default the value sets to 1.5

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=0.5);
```

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis='range');
```

### bootstrap

* Specifies whether to bootstrap the confidence intervals around the median for notched boxplots.
*  If bootstrap is None, no bootstrapping is performed, and notches are calculated using a Gaussian-based asymptotic  approximation
* Otherwise, bootstrap specifies the number of times to bootstrap(checking back) the median to determine its 95% confidence intervals. Values between 1000 and 10000 are recommended.

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= True,bootstrap=2000);
```

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= True,bootstrap=None);
```

### usermedians

* Overrides the median values
* Takes array or sequence as input
* Note dimension of sequence or array should match with firstt dimension of x
* If None, the median will be computed by matplotlib as normal


```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= True,bootstrap=5000,
            usermedians=[250,150,100]);
```

### conf_intervals

* Provides confidence Interval
* Takes array or sequence as input
* Note that first dimension of the input should match with x and second dimension must be  2
* If not None , the notch locations computed by matplotlib are ovverriden
* If None, the notches are computed by the method specified by the the other arguments e.g bootstrap


```python
a = np.random.randint(1,size=(3,2)).tolist()
```

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= True,bootstrap=5000,
           conf_intervals = a  );
```

### positions

* Sets the position of boxes(2,3,4)
* The ticks and limits are automatically set to match the position
* Default the range is (1,N+1)<br/>
  here, N is no of boxes to be drawn
* Takes array-like value

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= False,bootstrap=5000,
           positions = range(2, 5));
```

### widths

* Sets the width of each box
* Takes either scalar or sequence
* By default the value sets to 0.5


```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= False,bootstrap=5000,
           positions = range(1, 4),widths=[.8,.8,.8]);
```

### patch_artist

* If False produces boxes with the Line2D artist. Otherwise, boxes and drawn with Patch artists.
* By default the value is set to False

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis=10,notch= False,bootstrap=5000,
           positions = range(1, 4),widths=[.5,.5,.5], patch_artist = True);
```

### labels

* Takes sequence as input
* Note that sequence length mus be compitable with dimensions

### autorange

* If whis sets to 'range' such that the whisker ends are at the minimum and maximum of the data.

### meanline 

* If True (and showmeans is True), will try to render the mean as a line spanning the full width of the box according to meanprops (see below).
* Not recommended if shownotches is also True. Otherwise, means will be shown as points.

### capprops & boxprops & whiskerprops & flierprops & medianprops & meanprops

*  props are used to set style to specific elements leets say cap , box, whisker, or median line
* Takes dicts objects with styles specified.
* By default they takes None 
* Check this link for more info [styles](https://matplotlib.org/users/dflt_style_changes.html)

```python
plt.boxplot([data['TV'],data['radio'],data['newspaper']],sym=None,vert=True, whis='range',notch= False,bootstrap=5000,
           positions = range(1, 4),widths=[.5,.5,.5], patch_artist = True,labels=['TV','radio','newspaper'],manage_xticks=True,
           autorange = True,meanline = True,showmeans=True, meanprops={'linestyle':'--','color':'white'});
```
