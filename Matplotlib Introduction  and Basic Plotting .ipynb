## Legends, Title, Labels, Xticks,Yticks

### title  

* Set a title of the current axes
* Set one of the three available axes titles. The available titles are positioned above the axes in the center,flush with the     left edge, and flush with the right edge
* Parameters
     * label, Takes text to use for the title
     * fontdict, Takes a dictionary controlling the appearance of the title text, the default fontdict is:
        {‘fontsize’: rcParams[‘axes.titlesize’], ‘fontweight’ : rcParams[‘axes.titleweight’], ‘verticalalignment’:
        ‘baseline’, ‘horizontalalignment’: loc}
     * loc, Sets location of Title. Takes 'right' | 'left' | 'center', the default value is 'center'
      
### xlabel & ylabel
* Sets spacing in points between the label and the x-axis/y-axis
* Parametes,
     * xlabel and y label ,  Takes label text 
     * labelpad , Takes scalar, Default to None

### legend  

* Sets legends to plots

### xticks &  yticks 

* Get or set the current tick locations and labels of the x/y -axis
* Parameters,
     * locs : Takes a list of positions at which ticks should be placed. We can pass an empty list to disable xticks.
     * labels : Takes a list of explicit labels to place at the given locs.
     * ALong with above parameters text properties can be used to control the appearance of the labels
     * These function also returns loc and label values


** To set title ,labels,legend,ticks to pyplot **

```python
plt.plot(df['X'],df['X^2'])
plt.title('X vs X^2 Plot')
plt.xlabel('X')
plt.ylabel('X^2')
plt.legend()
plt.xticks([2,4,6,8])
plt.yticks([4,16,36,64])
plt.show()
```

** plotting data using  dataframe plot feature **

```python
df.plot(title='X vs X^2 Plot',legend=True,xticks=[2,4,6,8],yticks=[4,16,36,64],x='X',y='X^2')
```

```python
plt.plot(df['X'],df['X'],label= 'X line',linewidth=5)
plt.plot(df['X'],df['X^2'],label= 'X^2 line',linewidth=4)
plt.plot(df['X'],df['X^3'],label= 'X^3 line', linewidth=3)

plt.title('Plottting Powers of X')
plt.set_cmap('winter')
plt.show()
```

## Sub plotting

* Returns a subplot axes at the given grid position.
* Parameters
  * nrows and ncols 
     * Number of rows/columns of the subplot grid should has
     * Takes int value
     * by default the  value sets to 1
  * sharex, sharey : bool or {‘none’, ‘all’, ‘row’, ‘col’}, default: False
     * If True or all, x or y axis will be shared among all subplots
     * If False or none , each sublot will share an independent x or y axis
     * If row, each subplot row will share an x or y axis
     * If col, each subplot column will share an x or y axis
* When subplots have a shared x-axis along a column, only the x tick labels of the bottom subplot are visible
* when subplots have a shred y-axis along a row, only the y tick labels of the first column subplot are visible
 

** Pilotting (row = 1,col = 2) subplot where y axis is shared**

```python
fig, (ax1,ax2) = plt.subplots(nrows= 1, ncols=2, sharey=True)
ax1.plot(df['X'],df['X^2'])
ax1.set_title('Sharing Y axis')
ax2.scatter(df['X'], df['X^2'] );
```

** Pilotting (row = 2,col = 2) subplot where y axis is shared**

```python
fig, axes = plt.subplots(2, 2, sharey=True)
axes[0,0].plot(df['X'],df['X^2'])
axes[0,0].set_title('Sharing Y axis')
axes[0,1].scatter(df['X'], df['X^2'] )
axes[1,0].plot(df['X'], df['X^3'])
axes[1,1].scatter(df['X'], df['X^3']);
```

** Pilotting (row = 2,col = 2) subplot where y axis is not shared**

```python
fig, axes = plt.subplots(2, 2,sharex=False ,sharey=False)
axes[0,0].plot(df['X'],df['X^2'])
axes[0,0].set_title('Sharing Y axis')
axes[0,1].scatter(df['X'], df['X^2'] )
axes[1,0].plot(df['X'], df['X^3'])
axes[1,1].scatter(df['X'], df['X^3']);
```

** Pilotting (row = 2,col = 2) subplot where x axis is shared along columns**

```python
fig, axes = plt.subplots(2, 2, sharex = 'col')
axes[0,0].plot(df['X'],df['X^2'])
axes[0,0].set_title('Sharing Y axis')
axes[0,1].scatter(df['X'], df['X^2'] )
axes[1,0].plot(df['X'], df['X^3'])
axes[1,1].scatter(df['X'], df['X^3']);
```

** Pilotting (row = 2,col = 2) subplot where y axis is shared along row**

```python
fig, axes = plt.subplots(2, 2, sharey = 'row')
axes[0,0].plot(df['X'],df['X^2'])
axes[0,0].set_title('Sharing Y axis')
axes[0,1].scatter(df['X'], df['X^2'] )
axes[1,0].plot(df['X'], df['X^3'])
axes[1,1].scatter(df['X'], df['X^3']);
```
