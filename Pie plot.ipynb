# Pie Chart

* Plots pie chart of given x
* The fractional area of each wedge is given by x/sum(x). If sum(x) < 1, then the values of x give the fractional area directly   and the array will not be normalized. The resulting pie will have an empty wedge of size 1 - sum(x).
* The wedges are plotted counterclockwise, by default starting from the x-axis.



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Dataset

```python
import pandas as pd

states = ['Telengana', 'Bihar', 'West Bengal', 'Uttar Pradesh','Tripura']
companies = [100000, 25000, 175000, 75000,1000]

data = pd.DataFrame({'state':states,'companies':companies},columns=['state','companies'])
data
```

###  Basic Pie Chart Plotting using only x(data)

* Divides wedge sizes from givien x
* x takes array-line value

```python
plt.pie(x=data['companies']);
```

### labels

* Provides labels for each wedge
* Takes list of strings
* By default the value is set to None

```python
plt.pie(x=data['companies'],labels=data['state']);
```

### explode

* Takes array-like value which specifies the fraction of the radius and  length equals to x
* By default it is None
* If not None , then it offset each wedge as per value given in eplode array-ike object <br/>

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0]);
```

### colors

* Fills wedges with given color
* If None,  will use the colors in the currently active cycle.
* Takes array-like values
* By default the vlue sets to None

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b']);
```

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%');
```

### autopct

* If not None, provides numneric value to the wedge
* Takes string or function as values
* By default the value sets to None
*  If it is a format string, the label will be fmt%pct
* If it is a function , then fucntion will be called

```python
def autopact():
    return '%1.1f%%';

plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct=autopact());
```

### pctdistance

* Can configure the autopct label position
* Takes the float value which represents the ratio of the center of the pie slice and start of the text generated by autopct
* If autopct is None, then ignores the pctdistance 
* By deafult the value is set to 0.6


```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],
        autopct='%1.1f%%',pctdistance=1.5);
```

### shadow

* Draws a shadow beneath the pie chart
* Takes boolean value(True/False)
* By default the value is set to False

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],
        autopct='%1.1f%%',pctdistance=0.5,shadow=True);
```

### labeldistance

*  Can configure pie chart label positions
* Takes float value
* By default the value sets to 1.1

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=0.3);
```

### startangle

* If not None, rotates the start of the pie chart by the given agnle degrees counterclockwise from the x-axis
* Take int/float value
* By default the value sets to None

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90.0);
```

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=180.0);
```

### radius

* Can Specify the radius of the Pie chart
* Takes float value
* By default the value sets to None
* If radius is None, it will be set to 1

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1.5);
```

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=0.5);
```

### counterclock

* Specifies the fractions direction, clockwise or counterclockwise.
* Takes bools vlaue(True/False)
* By default the value sets to True

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1,counterclock=False);
```

### wedgeprops

* Can Pass dict of object to the wedge objects to make the pie
* Example, wedgeprops = {'linewidth': 3}
* For more info check this link: [wedge objects](https://matplotlib.org/api/_as_gen/matplotlib.patches.Wedge.html)
* By default the value sets to None

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1,counterclock=False,wedgeprops = {'linewidth': 10});
```

### textprops

* Takes dict of arguments to pass to the text objects.
* By default the value sets to None

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1,counterclock=False,wedgeprops = {'linewidth': 10},
       textprops={'text': 'I am wedge'});
```

### center

* Can privide center of the pie chart
* Takes sequence of 2 scalars
* By default the value sets to (0,0)

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1,counterclock=False,wedgeprops = {'linewidth': 10},
        center=(50,50));
```

### frame

* Plots axes frame with the chart if it sets to  True
* Takes bool vlue(True/False)
* By default the value sets to False

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1,counterclock=False,wedgeprops = {'linewidth': 10},
        center=(0,0),frame=True);
```

### rotatelabels

* Rotates each label to the angle of the corresponding slice ifit sets to True
* Takes bool values(True/False)
* By Default the value sets to False

```python
plt.pie(x=data['companies'],labels=data['state'], explode=[0,0.1,0,0,0],colors=['g','m','c','r','b'],autopct='%1.1f%%',
        pctdistance=0.5,shadow=True,labeldistance=1.1,startangle=90,radius=1,counterclock=False,wedgeprops = {'linewidth': 10},
        center=(0,0),frame=False,rotatelabels=True);
```

**Note** <br/>
In addition to the above described arguments, this function can take a data keyword argument. If such a data argument is given, the following arguments are replaced by data[<arg>]:<br/>
All arguments with the following names: colors, explode, labels, x
