# Scatter Plot

* A scatter plot of y vs x with varying marker size and/or color.<br/>
* Possible Parameters,<br/>
    matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, hold=None, data=None, **kwargs)


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Dataset 

```python
import pandas as pd
import numpy as np

data = pd.read_csv('../Advertising.csv',index_col= 0)
```

```python
data.head()
```

### Basic plotting using x and y parameters

* X and y are input data which are in array-like shape

```python
plt.scatter(data['sales'],data['TV'])
plt.scatter(data['sales'],data['radio'])
plt.scatter(data['sales'],data['newspaper'])

plt.legend()
plt.show()
```

### size

* Used as 's' param
* Sets size of the bubbles in scatter plot
* Takes scalar or array like values
* By default the value sets to rcParams['lines.markersize'] ** 2


```python
plt.scatter(data['sales'],data['TV'],s=10)
plt.scatter(data['sales'],data['radio'],s=20)
plt.scatter(data['sales'],data['newspaper'],s=30)

plt.legend()
plt.show()
```

### color, 

* Represents as 'c' param
* Takes sequence of colors
* By default the value sets to 'b'(blue)
* `c` can be a single color format string, or a sequence of color specifications of length `N`, or a sequence of `N` numbers to be mapped to colors using the `cmap` and `norm` specified via kwargs (see below).
*  Note that `c` should not be a single numeric RGB or RGBA sequence because that is indistinguishable from an array of values to be colormapped.  `c` can be a 2-D array in which the rows are RGB or RGBA, however, including the case of a single row to specify the same color for all points.
      

```python
plt.scatter(data['sales'],data['TV'],s=10,c='k')
plt.scatter(data['sales'],data['radio'],s=20,c='m')
plt.scatter(data['sales'],data['newspaper'],s=30,c='y')

plt.legend()
plt.show()
```

### marker

* Sets the shape for a data point
* Default the marker sets to  'o'
* Go through matplotlib.markers documnentation to check available markers in matplotlib
* marker can be either an instance of the class or the text shorthand for a particular marker.


```python
plt.scatter(data['sales'],data['TV'],s=10,c='k',marker='o')
plt.scatter(data['sales'],data['radio'],s=30,c='m',marker='+')
plt.scatter(data['sales'],data['newspaper'],s=40,c='y',marker='^')

plt.legend()
plt.show()
```

```python
plt.scatter(data['sales'],data['TV'],s=10,c='k',marker= r'$\clubsuit$')
plt.scatter(data['sales'],data['radio'],s=20,c='m',marker=r'$\heartsuit$')
plt.scatter(data['sales'],data['newspaper'],s=30,c='y',marker=r'$\diamondsuit$')

plt.legend()
plt.show()
```

### cmap

* Sets Regesterd colors to data points
* But cmap is only used if c is an array of floats. 
* If None, default to image.cmap

* See availabel color maps here: [colormaps](https://matplotlib.org/examples/color/colormaps_reference.html)

```python
plt.scatter(data['sales'],data['TV'],s=10,c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1')
plt.scatter(data['sales'],data['radio'],s=20,c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set1')
plt.scatter(data['sales'],data['newspaper'],s=30,c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set1')

plt.legend()
plt.show()
```

```python
plt.scatter(data['sales'],data['TV'],s=10,c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1')
plt.scatter(data['sales'],data['radio'],s=20,c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set2')
plt.scatter(data['sales'],data['newspaper'],s=30,c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set3')

plt.legend()
plt.show()
```

### norm

* Scales the data betweent 0, 1
* Takes boolean values
* But only used if c is an array of ploats 
* By default the value sets to None ,use the default :func:`normalize`.


```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1',norm=True)
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set2',norm=True)
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set3',norm=True)

plt.legend()
plt.show()
```

### vmin and vmax

* vmin and vmax are used in conjunction with norm to normalize data.
* Takes scalar value
* By default the values set to None
* If either are None, the min and max of the color array is used.
* PS : Note if you pass a norm instance, your settings for vmin and vmax will be ignored.
      

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1',vmin=0,vmax=1)
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set2',vmin=0,vmax=1)
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set3',vmin=0,vmax=1)

plt.legend()
plt.show()
```

### alpha

* Usful to configure transperency of data points
* Takes scalar value between 0(transparent), 1(opeque)
* By default the value sets to None

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1',vmin=0,vmax=1,alpha=0.5)
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set2',vmin=0,vmax=1,alpha=0.5)
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set3',vmin=0,vmax=1,alpha=0.5)

plt.legend()
plt.show()
```

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1',vmin=0,vmax=1,alpha=1)
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set2',vmin=0,vmax=1,alpha=1)
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set3',vmin=0,vmax=1,alpha=1)

plt.legend()
plt.show()
```

### linewidth

* Sets width of the line
* Takes scalar or array_like values
* By default the value sets to None

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),marker= r'$\clubsuit$',cmap='set1',vmin=0,vmax=1,alpha=0.5,linewidth=2)
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),marker=r'$\heartsuit$',cmap='set2',vmin=0,vmax=1,alpha=0.5,linewidth=5)
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),marker=r'$\diamondsuit$',cmap='set3',vmin=0,vmax=1,alpha=0.5,linewidth=4)

plt.legend()
plt.show()
```

### verts

*  If markeris None, these vertices will be used to construct the marker.
*  Takes sequenc of (x,y)
* The overall marker is rescaled by s
        

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),cmap='set1',vmin=0,vmax=1,alpha=0.5,linewidth=2,verts=(data[['sales','TV']]))
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),cmap='set2',vmin=0,vmax=1,alpha=0.5,linewidth=5,verts=(data[['sales','radio']]))
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),cmap='set3',vmin=0,vmax=1,alpha=0.5,linewidth=4,verts=(data[['sales','newspaper']]))

plt.legend()
plt.show()
```

### edgecolors

* Sets edge color
* Takes color or sequence of colors
* By default value sets to None<br/>
  If None, defaults to 'face'<br/>
  If 'face', the edge color will always be the same as the face color.
* If it is 'none', the patch boundary will not  be drawn.
*  For non-filled markers, the `edgecolors` kwarg is ignored and forced to 'face' internally. 
              
    
       

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),cmap='set1',vmin=0,vmax=1,alpha=0.5,linewidth=2,verts=(data[['sales','TV']]),edgecolors='g')
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),cmap='set2',vmin=0,vmax=1,alpha=0.5,linewidth=5,verts=(data[['sales','radio']]),edgecolors='m')
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),cmap='set3',vmin=0,vmax=1,alpha=0.5,linewidth=4,verts=(data[['sales','newspaper']]),edgecolors='c')

plt.legend()
plt.show()
```

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),cmap='set1',vmin=0,vmax=1,alpha=0.5,linewidth=2,verts=(data[['sales','TV']]),edgecolors=['g','m','c'])
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),cmap='set2',vmin=0,vmax=1,alpha=0.5,linewidth=5,verts=(data[['sales','radio']]),edgecolors=['k','black','violet'])
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),cmap='set3',vmin=0,vmax=1,alpha=0.5,linewidth=4,verts=(data[['sales','newspaper']]),edgecolors=['r','orange'])

plt.legend()
plt.show()
```

```python
plt.scatter(data['sales'],data['TV'],c=np.random.rand(3),cmap='set1',vmin=0,vmax=1,alpha=0.5,linewidth=2,verts=(data[['sales','TV']]),edgecolors='none')
plt.scatter(data['sales'],data['radio'],c=np.random.rand(3),cmap='set2',vmin=0,vmax=1,alpha=0.5,linewidth=5,verts=(data[['sales','radio']]),edgecolors='none')
plt.scatter(data['sales'],data['newspaper'],c=np.random.rand(3),cmap='set3',vmin=0,vmax=1,alpha=0.5,linewidth=4,verts=(data[['sales','newspaper']]),edgecolors='none')

plt.legend()
plt.show()
```

### Plot on DataFrame

* Basic plotting 
   * Takes x, y : label or position, optional

```python
data.plot.scatter(x = 'sales', y = 'TV');
```

* Optinal Parameters
    * s (size), Takes scalar or  array-like value
    * c (color), Takes label or array-like value

```python
data.plot.scatter(x = 'sales', y = 'TV', c = 'm',s = 25);
```
