## Hexabin Plotting 
* Used in bivarient analysis of high density dataset
* Makes a hexagonal binning plot of x versus y of the same length
* If C param is None, this is a histogram of number of occurrences of the observations ar (x[i],y[i]
* These values are accumulated for each hexagonal bin and then reduced according to reduce_C_function, which defaults to numpyâ€™s mean function (np.mean). (If C is specified, it must also be a 1-D sequence of the same length as x and y.)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Dataset

```python
import pandas as pd 
import numpy as np 
   
data = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
data['b'] = data['b'] + np.arange(1000)
data.head(10)
```

### Basic Plotting Using x and y Parameters

```python
plt.hexbin(x=data['a'],y=data['b']);
```

### C

* C parameter which represents the value of each point
* takes array or masked array
* Defaul the value sets to None

```python
plt.hexbin(x=data['a'],y=data['b'], C = data['a'] > 0 )
plt.scatter(data['a'],y=data['b'], c = data['a'] > 0,s= 0.5, cmap=plt.get_cmap('jet'));
```

```python
plt.hexbin(x=data['a'],y=data['b'], C = data['a'] >= 0 );
```

### gridsize

* Can configure the number of hexagonals in x- direction and y-direction
* If the value passed is scalar, y dirrection is chosen such that the hexagons are approximately regular.
* If gridsize is a tuple with two elements , then the elements specifies the number of hexagons in the x-direction and y-direction
* By default the value sets to 100

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25);
```

### bins

* Used to determine the color of the hexagon
* Takes 'log', int, sequence of values
* If None, no binning is applied; the color of each hexagon directly corresponds to its count value.
* If â€˜logâ€™, use a logarithmic scale for the color map. Internally,  is used to determine the hexagon color.
* If an integer, divide the counts in the specified number of bins, and color the hexagons accordingly.
* If a sequence of values, the values of the lower bound of the bins to be used.
* By default the value sets to None

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25, bins=3);
```

```python
bins = [0,1,3,2,4,5]
plt.hexbin(x=data['a'],y=data['b'], gridsize=25, bins=bins);
```

### xscale and yscale
* Use a linear or log10 scale on the horizontal and verical axis respectively.
* By default he value sets to log10

```python
plt.hexbin(x=data['a'],y=data['b'],gridsize= 25,xscale ='log10', yscale='log10');
```

### mincnt
* sets condition on display points 
* Avoids plotting hexagon for empty cells if mincnt = 1
* Takes positive integer value(int >0)
* By default the value sets to None
* If not None, only display cells with more than mincnt number of points in the cell

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=5);
```

### marginals

* If True, plot the marginal density as colormapped rectagles along the bottom of the x-axis and left of the y-axis
* Takes bool values 
* Default value is False

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3, marginals = True)
plt.scatter(data['a'],y=data['b'],s= 0.5, cmap=plt.get_cmap('jet'));
```

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3, marginals = True);
```

### extent

* Sets the limits on bins.
* Default assign the limits based on gridsize 
* Order of sclar is Order of scalars is (left, right, bottom, top)

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,extent=(-1, 2, 100, 1000));
```

### cmap

* Sets sequnce of colors to bins
* Takes matplotlib.colors.Colormap instance
* By defualt the value sets to None.
* If value is set to None it takes rc image.cmap vlue


```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,extent=(-1, 2, 100, 1000), cmap='Set1');
```

### alpha

* Can Adjust the transparency 
* Takes scalar value between 0, 1
* By Default the value sets to None

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,extent=(-1, 2, 100, 1000),alpha=0.5);
```

### linewidths

* Sets the line width 
* By defaul the value set to None
* If None, value sets to 1.0

```python
plt.hexbin(x=data['a'],y
           =data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,extent=(-1, 2, 100, 1000),
           alpha=0.5, linewidths=5);
```

### edgecolors
* Colors the edges of bins
* Takes  â€˜faceâ€™, â€˜noneâ€™, None or color 
* If â€˜faceâ€™ (the default), draws the edges in the same color as the fill color.
* If â€˜noneâ€™, no edge is drawn; this can sometimes lead to unsightly unpainted pixels between the hexagons.
* If None, draws outlines in the default color.
* If a matplotlib color arg, draws outlines in the specified color.

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,extent=(-1, 2, 100, 1000),
           alpha=0.5, linewidths=5,edgecolors=['m']);
```

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,extent=(-1, 2, 100, 1000),
           alpha=0.5, linewidths=5,edgecolors='none');
```

### vmin and vmax 

vmin and vmax are used in conjunction with norm to normalize luminance data. If None, the min and max of the color array C are used. Note if you pass a norm instance your settings for vmin and vmax will be ignored.

```python
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3,
           alpha=0.5, linewidths=5, vmin=1,vmax=5);
```

### norm

* Scales data to 0,1
* Takes object value
* By defalt the value sets to None

```python
from matplotlib import colors
plt.hexbin(x=data['a'],y=data['b'], gridsize=25,xscale='log2',yscale='log2', mincnt=3, cmap='Set1', 
           norm =  colors.Normalize(vmin=1, vmax=5, clip=False));
```
